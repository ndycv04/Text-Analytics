```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_page(soup,quotes):
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        quotes.append({'Text': text, 'Author':author})
        
base_url = 'https://quotes.toscrape.com'
headers = {'User-Agent': 'Mozilla/5.0'}

quotes = []

def scrape_all_pages(url):
    while url:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        scrape_page(soup, quotes)
        next_page = soup.find('li', class_='next')
        url = base_url + next_page.find('a')['href'] if next_page else None

scrape_all_pages(base_url)

einstein_quotes = []
for quote in quotes:
    if quote['Author'] == 'Albert Einstein':
        einstein_quotes.append(quote)

with open('quotes_Einstein.csv','w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text','Author'])
    writer.writeheader()
    writer.writerows(einstein_quotes)
```
